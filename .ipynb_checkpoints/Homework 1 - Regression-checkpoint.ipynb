```python
class NeuralNetwork:
    def __init__(self, input_size, hidden_layer_size=3, output_size=1):
        self.W1 = np.random.rand(input_size,hidden_layer_size) 
        self.W2 = np.random.rand(hidden_layer_size,output_size)
        
    def feedforward(self, X):
        self.layer1 = self.sigmoid(np.dot(X, self.W1))
        self.output = self.sigmoid(np.dot(self.layer1, self.W2))
        
        return self.output
        
    def backprop(self, X, y, y_hat):
        # backward propagate through the network
        self.o_error = y - y_hat # error in output
        self.o_delta = self.o_error*self.sigmoidPrime(y_hat) # applying derivative of sigmoid to error
        self.layer1_error = self.o_delta.dot(self.W2.T) # z2 error: how much our hidden layer weights contributed to output error
        self.layer1_delta = self.layer1_error*self.sigmoidPrime(self.layer1) # applying derivative of sigmoid to z2 error

        self.W1 += X.T.dot(self.layer1_delta) # adjusting first set (input --> hidden) weights
        self.W2 += self.layer1.T.dot(self.o_delta) # adjusting second set (hidden --> output) weights
        
    def sigmoid(self, s):
        # activation function
        return 1/(1+np.exp(-s))
    
    def sigmoidPrime(self, s):
        #derivative of sigmoid
        return s * (1 - s)
    
    def train(self, X, y, epochs=1000):
        for i in xrange(epochs):
            y_hat = self.feedforward(X)
            
            print "Predicted Output: \n" + str(y_hat) 
            print "Loss: \n" + str(np.mean(np.square(y - y_hat))) # mean sum squared loss
            print "\n"
            
            self.backprop(X, y, y_hat)
    
```

```python
y_train = np.array(train['Heating Load'].as_matrix().reshape(1, len(train['Heating Load']))) / 100
y_train = y_train.transpose()
X_train = np.array(train.drop(axis=1, columns=['Heating Load', 'Cooling Load']))
print y_train.shape
```

```python
NN = NeuralNetwork(X_train.shape[1], 5)
NN.train(X_train, y_train)
```
